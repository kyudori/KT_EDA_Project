```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score as r2
from math import sqrt
```

#### LinearRegression

```python
from sklearn.linear_model import LinearRegression

LR_reg1 = LinearRegression()
LR_reg1.fit(train_x1, train_y1)

LR_reg2 = LinearRegression()
LR_reg2.fit(train_x2, train_y2)
```

```python
y_pred_LR1 = LR_reg1.predict(test_x1)
y_pred_LR2 = LR_reg2.predict(test_x2)

print('LinearRegression')
print("Root Mean Squared Error : {0:.5f}".format(sqrt(mse(test_y1,y_pred_LR1))))
print("Root Mean Squared Error : {0:.5f}".format(sqrt(mse(test_y2,y_pred_LR2))))
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import ModelCheckpoint
```
